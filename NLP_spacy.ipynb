# Sentence Boundary Detection
This is the identification of sentences in a text

```python
for sent in doc.sents:
    print(sent)
```

```python
sentence1 = list(doc.sents)[0]
print(sentence1)
```

```python
token1 = sentence1[2]
print(token1)
```

```python
# returns a Verbatim text content
token1.text
```

```python
# The syntactic parent, or “governor”, of this token.
token1.head
```

```python
# The leftmost token of this token’s syntactic descendants
token1.left_edge
```

```python
# The rightmost token of this token’s syntactic descendants
token1.right_edge
```

```python
# returns the Named entity type
token1.ent_type
```

```python
# IOB code of named entity tag. “B” means the token begins an entity, 
# “I” means it is inside an entity, “O” means it is outside an entity, and "" means no entity tag is set.
token1.ent_iob_
```

```python
# Base form of the token, with no inflectional suffixes.
token1.lemma_
```

```python
# returns the verb of a word 
sentence1[12].lemma_
```

```python
# Morphological analysis
token1.morph
```

```python
sentence1[12].morph
```

```python
# Coarse-grained part-of-speech from the Universal POS tag set
token1.pos_
```

```python
# Syntactic dependency relation.
token1.dep_
```

```python
# Language of the parent document’s vocabulary
token1.lang_
```
